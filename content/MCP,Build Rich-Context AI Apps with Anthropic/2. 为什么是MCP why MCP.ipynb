## 一、为什么是MCP

我们可以便捷地使用最好的LLM模型，但是如果无法来连接外部数据和上下文，那就无法发挥它的最大作用。
MCP是一种开源的模型上下文协议，它对的大模型如何连接外部数据和协同工作进行了标准化。

MCP并不是一个全新的发现，而是参考网络应用程序通过REST与后端及其他系统通信那样，为大模型和外部数据及工具指定协议和通信标准等。
很多用MCP可以实现的功能一般也有例外的实现方式，但是通过MCP可以让不同模型之间或不同数据源之间的通信标准化，即使用同一种语言。
这就避免了仅仅因为模型或数据源的不同就要一次一次地做重复的构建工作，借鉴2016年微软开发的语言服务器协议（LSP），它对集成开发环境与特定语言工具的交互方式进行了标准化。站在“巨人”的肩膀上，MCP也是目标明确，为模型和外部数据及工具的交互建立标准。

## 三、MCP的架构

MCP与模型无关且完全开源，开发者可以自行构建也可以通过开源社区得到MCP的哪里。构建或使用与MCP兼容的应用程序并连接到许多不同的服务器山给，来获取我们需要的特定类型的数据访问权限。我们可以有用于数据存储的服务器，用于用户关系管理工具的服务器甚至是用于版本控制的服务器，直接通过自然语言就可以交互。



例如我们为Google Drive构建一个MCP服务器（或直接使用一个现成的）。无论是AI助手还是智能体抑或是桌面应用程序，只要它和MCP兼容，我们就可以发挥想象力，任意地创建数据访问方式。




## 四、MCP的发展及价值

对于不同的人群，MCP都是具备价值的。
对于应用开发者，使用极少的代码就可以连接到MCP服务器；对于API开发者，构建一次MCP服务就可以在各处使用；对于应用程序的用户来说，只需提供一个MCP服务器的URL就能把需要的数据导入应用程序中；对于大型的企业和组织来说，不必再浪费精力而是直接给不同的团队进行独立集成。




MCP的生态正在快速发展，来自开源社区和大型公司、初创团队等开发者均加入了构建生态的队伍中。支持MCP的软件开发工具SDK是由不同公司和人工智能开发者主导的，并且以多种不同的语言编写。与MCP兼容的应用程序已经涵盖网页应用、桌面应用、甚至智能体产品。



任何人都可以编写MCP服务器，也可以直接从开源社区获取需要的服务器。但是注意MCP不是API，它是构建在API之上的，如果你不想调用API，那么就可以使用自然语言让MCP完成这个任务。
